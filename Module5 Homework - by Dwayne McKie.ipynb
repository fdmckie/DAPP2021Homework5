### Exercise 1. 
Create a dictionary named pairs with key value pairs of:
    - 'hello' and 'fire', 
    - 3 and 'seven', 
    - 'strong' and 45. 

Then, print out the dictionary.

```python
#create a dictionary called "pairs"
pairs = {}

#add key value pairs to the dictionary
pairs = {"hello" : "fire", 3 : "seven", "strong" : 45}

#print the key value pairs in the dictionary
print(pairs)
```

### Exercise 3. 
Delete one item from <b><i>pairs</i></b> dictionary. Then, print deleted item and the dictionary.

```python
#delete a key value pair from the pairs dictionary. the popitem method will do this
#also, store the deleted key value pair from the pairs dictionary in the "pair" variable
pair = pairs.popitem()

#print the deleted key value pair from the "pairs" dictionary
print(pair)

#print the key value pairs in the dictionary
print(pairs)
```
